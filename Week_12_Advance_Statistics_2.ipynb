# **Advance Statistics - 2**


1. What is hypothesis testing in statistics?
  - Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves:
    * Making an initial assumption (called the null hypothesis).
    * Collecting sample data.
    * Using statistical tests to determine whether the data provides enough evidence to reject the initial assumption.

  - It helps answer questions like, “Is this drug effective?” or “Are the average scores of two classes different?”



2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
    1. Null Hypothesis (H₀): A statement that there is no effect or no difference. It represents the status quo or default assumption.

    2. Alternative Hypothesis (H₁ or Ha): A statement that contradicts the null hypothesis. It suggests that there is an effect or a difference.
  
  Example:
  H₀: The mean weight of apples = 150g
  H₁: The mean weight of apples ≠ 150g



3. What is the significance level in hypothesis testing, and why is it important?
  - The significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type I error).
    * Common values: 0.05, 0.01, 0.10
    * It defines the threshold for statistical significance.
  - If the p-value is less than α, we reject H₀.



4. What does a P-value represent in hypothesis testing?
  - The P-value is the probability of obtaining a test statistic as extreme or more extreme than the observed result, assuming the null hypothesis is true.
  - It helps determine how compatible your data is with the null hypothesis.



5. How do you interpret the P-value in hypothesis testing?
    * P-value ≤ α: Reject H₀ → The result is statistically significant.
    * P-value > α: Fail to reject H₀ → The result is not statistically significant.

  - Example: P = 0.03 and α = 0.05 → Reject H₀ → Significant evidence against H₀.



6. What are Type 1 and Type 2 errors in hypothesis testing?
  - Type I Error (False Positive): Rejecting H₀ when it is actually true         
    → Controlled by the significance level α.
  - Type II Error (False Negative): Failing to reject H₀ when H₁ is true.

    → Probability is denoted by β; Power = 1 − β.


7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
    * One-tailed test: Tests for a deviation in one direction (e.g., greater than).

    * Two-tailed test: Tests for a deviation in either direction (e.g., not equal to).

  Example:
      H₁: μ > 100 → one-tailed
      H₁: μ ≠ 100 → two-tailed
      


8. What is the Z-test, and when is it used in hypothesis testing?
  - A Z-test is used when:
      * The population standard deviation is known.
      * The sample size is large (n > 30).
      * The data is approximately normally distributed.
  - Used for comparing sample means to population means or proportions.



9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
  - Z-score formula (for a mean):
      Z = x̄ - μ / σ / √n

    Where:
     * x̄  = sample mean
     * μ = population mean
     * σ = population standard deviation
     * n = sample size

  - It tells how many standard deviations the sample mean is from the population mean.



10. What is the T-distribution, and when should it be used instead of the normal distribution?
  - The T-distribution is used when:
      * The sample size is small (n < 30).
      * The population standard deviation is unknown.
  - It is similar to the normal distribution but has heavier tails, allowing for more variability in small samples.



11. What is the difference between a Z-test and a T-test?
  1. Sample Size
     * Z-test: Typically used for large sample sizes (n > 30).
     * T-test: Preferred for small sample sizes (n ≤ 30).
  2. Population Standard Deviation Known or Unknown
     * Z-test: Used when the population standard deviation (σ) is known.
     * T-test: Used when the population standard deviation is unknown.
  3. Distribution Type
     * Z-test: Based on the standard normal distribution (mean = 0, SD = 1).
     * T-test: Based on the Student’s t-distribution, which is wider and has heavier tails than the normal distribution.


12. What is the T-test, and how is it used in hypothesis testing?
  - A T-test is used to compare means when the population standard deviation is unknown. Types:
     * One-sample t-test: Compare sample mean with population mean.
     * Two-sample t-test: Compare means of two independent groups.
     * Paired t-test: Compare means of the same group at two different times.



13. What is the relationship between Z-test and T-test in hypothesis testing?
  - Both tests are used to compare means. The T-test generalizes the Z-test for smaller samples or when σ is unknown.
     * As sample size increases, the T-distribution approximates the normal distribution.

14. What is a confidence interval, and how is it used to interpret statistical results?
  - A confidence interval (CI) provides a range of plausible values for a population parameter.
  - Example: A 95% CI means "we are 95% confident that the true population mean lies within this interval."
  - If the CI does not contain the null hypothesis value, we reject H₀.


15. What is the margin of error, and how does it affect the confidence interval?
  - The margin of error (ME) is the maximum expected difference between the sample statistic and the population parameter.

      CI = x̄ ± ME

    * Larger ME → Wider CI → Less precision.
    * Smaller ME → Narrower CI → More precision.

16. How is Bayes' Theorem used in statistics, and what is its significance?
  - Bayes’ Theorem updates the probability of a hypothesis based on new evidence.

  Bayes.jpg

  - Used in:
    * Medical testing
    * Spam filtering
    * Machine learning
  - It shifts thinking from frequentist to probabilistic reasoning.

17. What is the Chi-square distribution, and when is it used?
  - The Chi-square (χ²) distribution is used in tests of categorical data, especially:

    * Goodness of fit
    * Test of independence
    * Test for variance

  - It is skewed right and depends on degrees of freedom.

18. What is the Chi-square goodness of fit test, and how is it applied?
  - The Chi-square goodness of fit test is a statistical test used to determine whether a sample data matches a population with a specific distribution.
  - Steps:
    * Define expected frequencies.
    * Use formula:
    
    Chi-Square.jpg
    * Compare to critical value or use p-value.


19. What is the F-distribution, and when is it used in hypothesis testing?
  - The F-distribution is used in:
    * ANOVA
    * Comparing variances (F-test)
  - It is right-skewed and depends on two degrees of freedom: numerator and denominator.



20. What is an ANOVA test, and what are its assumptions?
  - ANOVA (Analysis of Variance) tests whether three or more group means are equal.
  - Assumptions:
    * Independence of observations
    * Normal distribution of groups
    * Equal variances (homogeneity)

21. What are the different types of ANOVA tests?
  - One-way ANOVA: One independent variable.
  - Two-way ANOVA: Two independent variables.
  - Repeated measures ANOVA: Same subjects tested multiple times.
  - MANOVA: Multiple dependent variables.

22. What is the F-test, and how does it relate to hypothesis testing?
  - The F-test compares two variances or is used in ANOVA to determine whether variances between group means are significantly different.
  - If F-statistic is large → Evidence against H₀ → At least one group mean is different.

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?
from scipy import stats
import numpy as np

# Sample data
sample = [5.1, 5.3, 5.2, 5.0, 5.4]
population_mean = 5.0
population_std = 0.2  # Known population std
n = len(sample)

# Z-test calculation
sample_mean = np.mean(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
# 2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python?
np.random.seed(0)
sample = np.random.normal(loc=102, scale=10, size=50)
population_mean = 100
population_std = 10

z_score = (np.mean(sample) - population_mean) / (population_std / np.sqrt(len(sample)))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z = {z_score:.2f}, P-value = {p_value:.4f}")
```

```python
# 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean?
def one_sample_z_test(sample, population_mean, population_std):
    n = len(sample)
    sample_mean = np.mean(sample)
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    p = 2 * (1 - stats.norm.cdf(abs(z)))
    return z, p

sample = np.random.normal(50, 5, 30)
z, p = one_sample_z_test(sample, population_mean=52, population_std=5)
print(f"Z = {z:.2f}, P = {p:.4f}")
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot?
import matplotlib.pyplot as plt

z, p = one_sample_z_test(sample, 52, 5)

# Plot
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)
plt.plot(x, y, label='Normal Distribution')
plt.fill_between(x, y, where=(x <= -1.96) | (x >= 1.96), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(z, color='blue', linestyle='--', label=f'Z = {z:.2f}')
plt.title("Two-Tailed Z-test")
plt.legend()
plt.show()
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?
def visualize_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    se = sigma / np.sqrt(n)
    x = np.linspace(80, 120, 1000)
    z_critical = stats.norm.ppf(1 - alpha/2)
    critical_value = mu0 + z_critical * se

    # Distributions under H0 and H1
    y0 = stats.norm.pdf(x, mu0, se)
    y1 = stats.norm.pdf(x, mu1, se)

    plt.plot(x, y0, label='H0', color='blue')
    plt.plot(x, y1, label='H1', color='orange')

    # Shade regions
    plt.fill_between(x, y0, where=(x > critical_value), color='red', alpha=0.4, label='Type I Error')
    plt.fill_between(x, y1, where=(x < critical_value), color='green', alpha=0.4, label='Type II Error')

    plt.legend()
    plt.title("Type I & II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.show()

visualize_errors()
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance?
def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    se = stats.sem(sample)
    margin = se * stats.t.ppf((1 + confidence) / 2, n - 1)
    return mean - margin, mean + margin

sample = np.random.normal(50, 5, 40)
ci = confidence_interval(sample)
print(f"95% Confidence Interval: {ci}")
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data?
def margin_of_error(sample, confidence=0.95):
    se = stats.sem(sample)
    moe = se * stats.t.ppf((1 + confidence) / 2, len(sample) - 1)
    return moe

moe = margin_of_error(sample)
print(f"Margin of Error: {moe:.2f}")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?
def bayes_theorem(prior_A, prob_B_given_A, prob_B_given_not_A):
    prob_not_A = 1 - prior_A
    prob_B = prob_B_given_A * prior_A + prob_B_given_not_A * prob_not_A
    posterior = (prob_B_given_A * prior_A) / prob_B
    return posterior

posterior = bayes_theorem(0.01, 0.95, 0.05)
print(f"Posterior probability: {posterior:.4f}")
```
