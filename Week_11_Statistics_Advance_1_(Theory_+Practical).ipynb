**1. What is a random variable in probability theory?**
  - A random variable is a numerical quantity whose value depends on the    outcome of a random experiment.

**2. What are the types of random variables?**
  - Types of Random Variables:
    1. Discrete Random Variable – takes a countable number of values (e.g., number of heads in 3 coin tosses).
    2. Continuous Random Variable – takes an infinite number of values over an interval (e.g., height, temperature).

**3. What is the difference between discrete and continuous distributions?**
  1. Nature of Values:
  - Discrete distributions deal with countable outcomes (e.g., number of students, dice rolls).

  - Continuous distributions involve uncountable, infinite outcomes within an interval (e.g., height, weight, time).

  2. Probability Assignment:
  - In discrete, you can assign a specific probability to each individual outcome (e.g., P(X = 3) = 0.2).

  - In continuous, the probability of any exact value is zero; instead, you calculate probability over a range (e.g., P(2 < X < 4)).

  3. Functions Used:
  - Discrete uses the Probability Mass Function (PMF).
  - Continuous uses the Probability Density Function (PDF).

  4. Graph Appearance:
  - Discrete distribution graphs appear as separate spikes or bars.
  - Continuous distribution graphs are smooth curves.

**4. What are probability distribution functions (PDF)?**
  - A probability distribution function describes how probability is distributed over the values of a random variable.

**5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**
  1. Definition:
  - PDF/PMF gives the probability of a random variable taking on a specific value.
  - CDF gives the probability that the variable is less than or equal to a certain value.

  2. Output Type:
  - PDF/PMF gives individual probabilities (e.g., P(X = x)).
  - CDF gives a cumulative probability up to a point (e.g., P(X ≤ x)).

  3. Usage:
  - PDF/PMF is used to find the probability at a single point.
  - CDF is used to calculate the probability within a range or below a threshold.

  4. Nature of Function:
  - PDF (for continuous) may return a density, not actual probability.
  - CDF always returns a value between 0 and 1, and it is non-decreasing.

**9. What is the Poisson distribution and where is it applied?**

  - Applications: Call arrivals, system failures, etc.

**10. What is a continuous uniform distribution?**

**11. What are the characteristics of a normal distribution?**

  - Defined by mean (μ) and standard deviation (σ)
  - Total area under curve = 1
  - Empirical Rule:

**13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**
  - CLT states: Regardless of the population distribution, the sampling distribution of the sample mean approaches a normal distribution as sample size increases.
      * Important for making inferences about the population using sample data

  - As per CLT, even if the population is non-normal, the sampling distribution of the mean becomes normal for large sample sizes (n > 30 typically).
  - Enables use of Z and t distributions.

**16. How do you calculate a Z-score, and what does it represent?**
  - A Z-score indicates how many standard deviations a data point is from the mean.
  - Z = X−μ / σ​
      * Z > 0 → above mean
      * Z < 0 → below mean.

**17. What are point estimates and interval estimates in statistics?**
  1. Point Estimates
  - A point estimate is a single value (statistic) used as the best guess for a population parameter.

  - Example:
  - Sample mean ( x̄ ) is a point estimate of the population mean (𝜇).
  - Sample proportion ( p̂ ) is a point estimate of the population proportion (p).

  2. Interval Estimate
  - An interval estimate gives a range of values (an interval) within which the population parameter is expected to lie, along with a certain level of confidence (usually 95% or 99%).
  - This is known as a confidence interval (CI).
  - Example:
  - We are 95% confident that the population mean lies between 13.5 and 16.2.


**18. What is the significance of confidence intervals in statistical analysis?**
  - Provide range of values likely to contain the population parameter
  - More informative than a point estimate
  - A 95% CI means: If we repeated the sampling, 95% of those intervals would contain the true mean

**19. What is the relationship between a Z-score and a confidence interval?**
  - Z- Score
  - A Z-score tells you how many standard deviations a data point is from the mean.
#  - **Formula:**
#  - **Z = X−μ / σ​**
      * X: a data value or sample mean
      * μ: population mean
      * σ: population standard deviation
  - Confidence Interval (CI):
  - A confidence interval gives a range of values that is likely to contain the true population parameter (like the mean), based on sample data and a certain confidence level (like 95%).

  - **Relationship Between Z-Score and Confidence Interval**
  - When estimating the population mean using a normal distribution (or when sample size is large enough for Central Limit Theorem to apply), the confidence interval uses critical Z-scores that correspond to the desired confidence level.


**20. How are Z-scores used to compare different distributions?**
  - Standardize different datasets for comparison
  - Example: Comparing scores from two different exams
      * Z-score tells how relatively high or low a score is

**21. What are the assumptions for applying the Central Limit Theorem?**
  - Independent observations.
  - Identically distributed samples.
  - Finite mean and variance.
  - Larger sample sizes (n ≥ 30) improve approximation.

**22. What is the concept of expected value in a probability distribution?**
  - The expected value is the mean of a random variable. It represents the long-run average outcome

**23. How does a probability distribution relate to the expected outcome of a random variable?**

  - A probability distribution shows how likely each outcome is. The expected value summarizes the average of these outcomes, weighted by their probabilities.

```python
# 2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define range
low, high = 1, 7  # Simulate a dice roll: values from 1 to 6

# Generate data
data = randint.rvs(low, high, size=1000)

# PMF plot
x = np.arange(low, high)
pmf = randint.pmf(x, low, high)

# plt.stem(x, pmf, basefmt=" ", use_line_collection=True)
plt.stem(x, pmf, basefmt=" ")
plt.title("PMF of Discrete Uniform Distribution (Dice Roll)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?
from scipy.stats import bernoulli

def bernoulli_pdf(p, x):
    return bernoulli.pmf(x, p)

# Example usage
p = 0.7
print("P(X=1):", bernoulli_pdf(p, 1))
print("P(X=0):", bernoulli_pdf(p, 0))
```

```python
# 4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram?
n, p = 10, 0.5
binomial_data = np.random.binomial(n, p, 1000)

plt.hist(binomial_data, bins=range(n+2), density=True, edgecolor='black', align='left')
plt.title("Binomial Distribution Histogram (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# 5. Create a Poisson distribution and visualize it using Python?
from scipy.stats import poisson

mu = 4  # Mean rate
x = np.arange(0, 15)
pmf = poisson.pmf(x, mu)

plt.stem(x, pmf, basefmt=" ")
plt.title("Poisson Distribution PMF (λ=4)")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

```python
# 6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution?
from scipy.stats import randint

low, high = 1, 7
x = np.arange(low, high)
cdf = randint.cdf(x, low, high)

plt.step(x, cdf, where='post')
plt.title("CDF of Discrete Uniform Distribution")
plt.xlabel("x")
plt.ylabel("CDF")
plt.grid(True)
plt.show()
```

```python
# 14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution?
def compute_z_score(x, mean, std_dev):
    return (x - mean) / std_dev

z = compute_z_score(85, mean=75, std_dev=5)
print(f"Z-score: {z}")
```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset?
from scipy.stats import norm

# Example values
sample_mean = 102
pop_mean = 100
std_dev = 10
n = 50

z = (sample_mean - pop_mean) / (std_dev / np.sqrt(n))
p_value = 1 - norm.cdf(z)

print(f"Z-score: {z:.3f}")
print(f"P-value: {p_value:.4f}")
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result?
import scipy.stats as stats

data = np.random.normal(50, 10, 100)
mean = np.mean(data)
sem = stats.sem(data)
confidence = 0.95
margin = stats.t.ppf((1 + confidence) / 2.0, len(data)-1) * sem

print(f"Confidence Interval: ({mean - margin:.2f}, {mean + margin:.2f})")
```

```python
# 17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean?
data = np.random.normal(70, 15, 200)
mean = np.mean(data)
sem = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

print(f"95% Confidence Interval for Mean: {ci}")
```
